```python
zeros = [0] * len(uwords)
dwords = dict(zip(uwords, zeros))
```

```python
f = open('thehunterswife.txt', 'r')
paras = f.readlines()
f.close()
```

```python
paras
```

```python
for i in range(len(paras)):
    print(paras[i])
```

```python
a = paras[0]
```

```python
a
```

```python
c = []
b = a.split()
for w in range(len(b)-1):
    c.append((b[w], b[w+1]))
```

```python
print(c)
```

```python
bigram = []
for i in range(len(paras)):
    para = paras[i].lower().split()
    for j in range(len(para)-1):
        bigram.append((para[j], para[j+1]))
bigram
```

```python
len(bigram)
```

```python
ubigram = set(bigram)
```

```python
len(ubigram)
```

```python
zeros = [0] * len(ubigram)
dbigram = dict(zip(ubigram, zeros))
```

```python
for i in bigram:
    dbigram[i] = dbigram[i] + 1
dbigram
```
