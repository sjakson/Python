```python
import nltk
from bs4 import BeautifulSoup
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import PunktSentenceTokenizer

with open('Machine learning and medicine book review and commentary.html', 'rb') as html:
    soup = BeautifulSoup(html, features="html5lib")

text = soup.find('article')
text = text.get_text()

text_tokenized = word_tokenize(text)

customer_sent_tokenizer = PunktSentenceTokenizer(text)
tokenized = customer_sent_tokenizer.tokenize(text)

stop_words = set(stopwords.words("english"))

filtered_sentence = []

for w in text_tokenized:
    if w not in stop_words:
        filtered_sentence.append(w)

print(word_tokenize(text))
print(filtered_sentence)

for i in tokenized:
    words = nltk.word_tokenize(i)
    tagged = nltk.pos_tag(words)
    print(tagged)

    namedEnt = nltk.ne_chunk(tagged, binary=True)

    print(namedEnt)

    

```
